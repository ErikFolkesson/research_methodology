# 1: Initial Exploration

## 1.1.1: Simple bar plot of Prog-1 vs Prog-2

```python
# Extract Prog-1 and Prog-2 times
prog1_times = [time_data[prog]['Prog-1'] for prog in time_data]
prog2_times = [time_data[prog]['Prog-2'] for prog in time_data]

# Calculate the difference between Prog-1 and Prog-2
time_diffs = [prog1_times[i] - prog2_times[i] for i in range(len(prog1_times))]
```

```python
# 2. Visualization: Prog-1 in positive direction, Prog-2 in negative, and the difference as a line graph
plt.figure(figsize=(12, 6))

# Plot Prog-1 in the positive direction
plt.bar(time_data.keys(), prog1_times, label='Prog-1', color='blue', alpha=0.6)

# Plot Prog-2 in the negative direction
plt.bar(time_data.keys(), [-t for t in prog2_times], label='Prog-2', color='green', alpha=0.6)

# Plot the difference between Prog-1 and Prog-2 as a line graph
plt.plot(time_data.keys(), time_diffs, label='Difference (Prog-1 - Prog-2)', color='red', marker='o', linewidth=4)

# Add labels and title
plt.xlabel('Programmer')
plt.ylabel('Time (minutes)')
plt.title('Time to Develop Prog-1 and Prog-2 for Each Programmer and Difference')

# Ensure x-axis shows programmer numbers from 1 to 10
plt.xticks(list(time_data.keys()), [str(prog) for prog in time_data.keys()])

plt.legend()
plt.show()
```

## 2: Parametric Tests
Is the data normally distributed?

### 2.1: IDE Data

```python
# Import necessary libraries
from scipy.stats import shapiro, kstest, anderson
import scipy.stats as stats
import matplotlib.pyplot as plt

# Shapiro-Wilk Test for IDE-A times
shapiro_test_ide_a = shapiro(ide_a_times)
print(f"Shapiro-Wilk Test for IDE-A: p-value = {shapiro_test_ide_a.pvalue:.4f}")

# Shapiro-Wilk Test for IDE-B times
shapiro_test_ide_b = shapiro(ide_b_times)
print(f"Shapiro-Wilk Test for IDE-B: p-value = {shapiro_test_ide_b.pvalue:.4f}")

# Anderson-Darling Test for IDE-A times
ad_test_ide_a = anderson(ide_a_times, dist='norm')
print(f"Anderson-Darling Test for IDE-A: statistic = {ad_test_ide_a.statistic:.4f}")
print("Critical values:", ad_test_ide_a.critical_values)
print("Significance levels:", ad_test_ide_a.significance_level)

# Anderson-Darling Test for IDE-B times
ad_test_ide_b = anderson(ide_b_times, dist='norm')
print(f"Anderson-Darling Test for IDE-B: statistic = {ad_test_ide_b.statistic:.4f}")
print("Critical values:", ad_test_ide_b.critical_values)
print("Significance levels:", ad_test_ide_b.significance_level)
```

### 2.2: Program Data

```python
# Shapiro-Wilk Test for Prog-1 times
shapiro_test_prog1 = shapiro(prog1_times)
print(f"Shapiro-Wilk Test for Prog-1: p-value = {shapiro_test_prog1.pvalue:.4f}")

# Shapiro-Wilk Test for Prog-2 times
shapiro_test_prog2 = shapiro(prog2_times)
print(f"Shapiro-Wilk Test for Prog-2: p-value = {shapiro_test_prog2.pvalue:.4f}")

# Anderson-Darling Test for Prog-1 times
ad_test_prog1 = anderson(prog1_times, dist='norm')
print(f"Anderson-Darling Test for Prog-1: statistic = {ad_test_prog1.statistic:.4f}")
print("Critical values:", ad_test_prog1.critical_values)
print("Significance levels:", ad_test_prog1.significance_level)

# Anderson-Darling Test for Prog-2 times
ad_test_prog2 = anderson(prog2_times, dist='norm')
print(f"Anderson-Darling Test for Prog-2: statistic = {ad_test_prog2.statistic:.4f}")
print("Critical values:", ad_test_prog2.critical_values)
print("Significance levels:", ad_test_prog2.significance_level)
```

### 2.3: IDE Effect on Each Program

```python
# Shapiro-Wilk Test for Prog-1 with IDE-A times
shapiro_test_prog1_ide_a = shapiro(prog1_ide_a_times)
print(f"Shapiro-Wilk Test for Prog-1 with IDE-A: p-value = {shapiro_test_prog1_ide_a.pvalue:.4f}")

# Shapiro-Wilk Test for Prog-1 with IDE-B times
shapiro_test_prog1_ide_b = shapiro(prog1_ide_b_times)
print(f"Shapiro-Wilk Test for Prog-1 with IDE-B: p-value = {shapiro_test_prog1_ide_b.pvalue:.4f}")

# Shapiro-Wilk Test for Prog-2 with IDE-A times
shapiro_test_prog2_ide_a = shapiro(prog2_ide_a_times)
print(f"Shapiro-Wilk Test for Prog-2 with IDE-A: p-value = {shapiro_test_prog2_ide_a.pvalue:.4f}")

# Shapiro-Wilk Test for Prog-2 with IDE-B times
shapiro_test_prog2_ide_b = shapiro(prog2_ide_b_times)
print(f"Shapiro-Wilk Test for Prog-2 with IDE-B: p-value = {shapiro_test_prog2_ide_b.pvalue:.4f}")

# Anderson-Darling Test for Prog-1 with IDE-A times
ad_test_prog1_ide_a = anderson(prog1_ide_a_times, dist='norm')
print(f"Anderson-Darling Test for Prog-1 with IDE-A: statistic = {ad_test_prog1_ide_a.statistic:.4f}")
print("Critical values:", ad_test_prog1_ide_a.critical_values)
print("Significance levels:", ad_test_prog1_ide_a.significance_level)

# Anderson-Darling Test for Prog-1 with IDE-B times
ad_test_prog1_ide_b = anderson(prog1_ide_b_times, dist='norm')
print(f"Anderson-Darling Test for Prog-1 with IDE-B: statistic = {ad_test_prog1_ide_b.statistic:.4f}")
print("Critical values:", ad_test_prog1_ide_b.critical_values)
print("Significance levels:", ad_test_prog1_ide_b.significance_level)

# Anderson-Darling Test for Prog-2 with IDE-A times
ad_test_prog2_ide_a = anderson(prog2_ide_a_times, dist='norm')
print(f"Anderson-Darling Test for Prog-2 with IDE-A: statistic = {ad_test_prog2_ide_a.statistic:.4f}")
print("Critical values:", ad_test_prog2_ide_a.critical_values)
print("Significance levels:", ad_test_prog2_ide_a.significance_level)

# Anderson-Darling Test for Prog-2 with IDE-B times
ad_test_prog2_ide_b = anderson(prog2_ide_b_times, dist='norm')
print(f"Anderson-Darling Test for Prog-2 with IDE-B: statistic = {ad_test_prog2_ide_b.statistic:.4f}")
print("Critical values:", ad_test_prog2_ide_b.critical_values)
print("Significance levels:", ad_test_prog2_ide_b.significance_level)
```

### 2.4: Conclusion
In conclusion we determine that the data is normally distributed. We can now proceed with parametric tests.

## 3: Parametric Tests

### 3.1: Difference Between Dev Time of Prog-1 and Prog-2
We can use a paired t-test to compare the development times of Prog-1 and Prog-2.

```python
from scipy.stats import ttest_rel

# Paired t-test for Prog-1 and Prog-2 times
t_statistic, p_value_prog1_vs_prog2 = ttest_rel(prog1_times, prog2_times)
print(f"Paired t-test for Prog-1 vs Prog-2: p-value = {p_value_prog1_vs_prog2:.4f}")
```

### 3.2: Difference Between IDE-A and IDE-B Dev Time
We can use a paired t-test to compare the development times for IDE-A and IDE-B.

```python
# Paired t-test for IDE-A and IDE-B times
t_statistic, p_value_ide_a_vs_ide_b = ttest_rel(ide_a_times, ide_b_times)
print(f"Paired t-test for IDE-A vs IDE-B: p-value = {p_value_ide_a_vs_ide_b:.4f}")
```

### 3.3: Difference Between Prog-1 Using IDE-A and IDE-B and Prog-2 Using IDE-A or IDE-B
We can use a two-way ANOVA to compare the development times for Prog-1 and Prog-2 using IDE-A and IDE-B.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Prepare data for two-way ANOVA
anova_data = pd.DataFrame({
    'Time': prog1_ide_a_times + prog1_ide_b_times + prog2_ide_a_times + prog2_ide_b_times,
    'Program': ['Prog-1'] * (len(prog1_ide_a_times) + len(prog1_ide_b_times)) + ['Prog-2'] * (len(prog2_ide_a_times) + len(prog2_ide_b_times)),
    'IDE': ['IDE-A'] * len(prog1_ide_a_times) + ['IDE-B'] * len(prog1_ide_b_times) + ['IDE-A'] * len(prog2_ide_a_times) + ['IDE-B'] * len(prog2_ide_b_times)
})

# Fit the two-way ANOVA model
model = ols('Time ~ C(Program) + C(IDE) + C(Program):C(IDE)', data=anova_data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# Perform paired t-tests
t_stat_prog1, p_value_prog1 = ttest_rel(prog1_ide_a_times, prog1_ide_b_times)
t_stat_prog2, p_value_prog2 = ttest_rel(prog2_ide_a_times, prog2_ide_b_times)

print(f"Paired t-test p-value for Prog-1: {p_value_prog1:.4f}")
print(f"Paired t-test p-value for Prog-2: {p_value_prog2:.4f}")
```
